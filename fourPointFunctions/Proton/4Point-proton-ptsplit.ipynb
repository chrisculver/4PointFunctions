```python
%load_ext autoreload
%autoreload 2

import sys
import os

sys.path.append(os.getcwd()[:-7])

import WickContractions

# the uglies part of the code is still defining operators.
from WickContractions.ops.operator import Operator
from WickContractions.ops.elemental import ElementalOperator
from WickContractions.ops.quarks import Quark
from WickContractions.ops.indexed import IndexedObject

#pretty latex printing
from IPython.display import display, Math 
pprint = lambda o : display(Math(str(o)))
```

```python
# epsilion^abc (u_a Cg5 d_b^T) u_c
q1=Quark(False,'u','s_0','a','t_3','x_3')
g=IndexedObject('\\Gamma',['s_0','s_1'])
eps=IndexedObject('\\epsilon',['a','b','c'])
q2=Quark(False,'d','s_1','b','t_3','x_3')
q3=Quark(False,'u','s_2','c','t_3','x_3')

annihilate=Operator([ElementalOperator(+1,[g,eps],[q1,q2,q3])])


# epsilion^abc (baru_a Cg5 bard_b^T) baru_c
elementals=[]
for pi2 in [['u',+2/3],['d',-1/3]]:
    for pi1 in [['u',+2/3],['d',-1/3]]:
        for dirpi2 in [-1,+1]:                               # eq 17 of notes
            for dirpi1 in [-1,+1]:
                q4=Quark(True,pi2[0],'s_3','c_1','t_2','x_2')
                g1=IndexedObject('((1-\\gamma_{\mu})e^{-i\\mathbf{q}})',['s_3','s_4'])
                g2=IndexedObject('U_{\mu}(t_2,t_2+\hat{\mu_4})',['c_1','c_2'])
                q5=Quark(False,pi2[0],'s_4','c_2','t_2','x_2+\\hat{\mu}')

                q6=Quark(True,pi1[0],'s_5','c_3','t_1','x_1')
                g3=IndexedObject('((1-\\gamma_{\\nu})e^{i\\mathbf{q}})',['s_5','s_6'])
                g4=IndexedObject('U_{\\nu}(t_1,t_1+\hat{\nu_4})',['c_3','c_4'])
                q7=Quark(False,pi1[0],'s_6','c_4','t_1','x_1+\\hat{\\nu}')

                if(dirpi1==-1):
                    q4=Quark(True,pi2[0],'s_3','c_1','t_2','x_2')
                    g1=IndexedObject('((1-\\gamma_{\mu})e^{-i\\mathbf{q}})',['s_3','s_4'])
                    g2=IndexedObject('U_{\mu}(t_2,t_2+\hat{\mu_4})',['c_1','c_2'])
                    q5=Quark(False,pi2[0],'s_4','c_2','t_2','x_2+\\hat{\mu}')
                elif(dirpi1==+1):
                    q4=Quark(True,pi2[0],'s_3','c_1','t_2','x_2+\\hat{\mu}')
                    g1=IndexedObject('((1+\\gamma_{\mu})e^{-i\\mathbf{q}})',['s_3','s_4'])
                    g2=IndexedObject('U_{\mu}^{\dagger}(t_2+\hat{\mu_4},t_2)',['c_1','c_2'])
                    q5=Quark(False,pi2[0],'s_4','c_2','t_2','x_2')
    
                if(dirpi2==-1):
                    q6=Quark(True,pi1[0],'s_5','c_3','t_1','x_1')
                    g3=IndexedObject('((1-\\gamma_{\\nu})e^{i\\mathbf{q}})',['s_5','s_6'])
                    g4=IndexedObject('U_{\\nu}(t_1,t_1+\hat{\nu_4})',['c_3','c_4'])
                    q7=Quark(False,pi1[0],'s_6','c_4','t_1','x_1+\\hat{\\nu}')
                    
                elif(dirpi2==+1):
                    q6=Quark(True,pi1[0],'s_5','c_3','t_1','x_1+\\hat{\\nu}')
                    g3=IndexedObject('((1+\\gamma_{\\nu})e^{i\\mathbf{q}})',['s_5','s_6'])
                    g4=IndexedObject('U_{\\nu}^{\dagger}(t_1+\hat{\nu_4},t_1)',['c_3','c_4'])
                    q7=Quark(False,pi1[0],'s_6','c_4','t_1','x_1')

                # the last proton
                pq1=Quark(True,'u','s_7','a\'','t_0','x_0')
                pg=IndexedObject('\\Gamma',['s_7','s_8'])
                peps=IndexedObject('\\epsilon',['a\'','b\'','c\''])
                pq2=Quark(True,'d','s_8','b\'','t_0','x_0')
                pq3=Quark(True,'u','s_9','c\'','t_0','x_0')
                
                coef=pi1[1]*pi2[1]*dirpi1*dirpi2

                elementals.append(ElementalOperator(-1*coef,[g1,g2,g3,g4,pg,peps],[q4,q5,q6,q7,pq1,pq2,pq3]))
        
create=Operator(elementals)
```
